#### Q3. Here we explore the maximal margin classifier on a toy data set.

```
from sklearn import svm
```

```
# defining data
X = [[3, 2, 4, 1, 2, 4, 4], [4, 2, 4, 4, 1, 3, 1]]
y = ['Red', 'Red', 'Red', 'Red', 'Blue', 'Blue', 'Blue']
```

```
# a)

plt.figure(figsize=(8, 6))

sns.scatterplot(X[0], X[1], hue=y, ci=None, palette=["red", "blue"], s=80)

plt.title("Q3 (a)", color='black', fontsize=18)
plt.legend(facecolor='black')
plt.show()
```

```
# b)
clf = svm.SVC(kernel='linear', C=100)
clf.fit(np.array(X).T.tolist(), y)

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 5)
yy = a * xx - (clf.intercept_[0] / w[1])

plt.figure(figsize=(8, 6))

sns.scatterplot(X[0], X[1], hue=y, ci=None, palette=["red", "blue"], s=80)

plt.plot(xx, yy, 'k-')

plt.title("Q3 (b)", color='black', fontsize=18)
plt.legend(facecolor='black')
plt.show()
```

Eqn for hyperplane:

$$0.99970703 - 1.99941406 X_1 + 1.99941406 X_2 = 0$$

c)

Classify to Red < 0 and Blue otherwise.

```
# d)
clf = svm.SVC(kernel='linear', C=100)
clf.fit(np.array(X).T.tolist(), y)

# sep hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 5)
yy = a * xx - (clf.intercept_[0] / w[1])

# margins
b = clf.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = clf.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])

plt.figure(figsize=(8, 6))

sns.scatterplot(X[0], X[1], hue=y, ci=None, palette=["red", "blue"], s=80)

plt.plot(xx, yy, 'k-')
plt.plot(xx, yy_down, 'k--', linewidth=3)
plt.plot(xx, yy_up, 'k--', linewidth=3)

plt.title("Q3 (d)", color='black', fontsize=18)
plt.legend(facecolor='black')
plt.show()
```

e)

Support vectors for maximal margin classier = $(2,1)$, $(2,2)$, $(4,3)$, $(4,4)$

f)

A slight movement of point $(4,1)$ would not affect the maximal margin hyperplane as hyperplanes are dependent on support vectors. Since point $(4,1)$ is "off the street" point, it won't affect the hyperplane.

```
# g)
clf = svm.SVC(kernel='linear', C=100)
clf.fit(np.array(X).T.tolist(), y)

# sep hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 5)
yy = a * xx - (clf.intercept_[0] / w[1])

# margins
b = clf.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = clf.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])

plt.figure(figsize=(8, 6))

sns.scatterplot(X[0], X[1], hue=y, ci=None, palette=["red", "blue"], s=80)

plt.plot(xx, yy, 'k-')
plt.plot(xx, xx - 0.3, 'k-', color='orange')
plt.plot(xx, yy_down, 'k--', linewidth=3)
plt.plot(xx, yy_up, 'k--', linewidth=3)

plt.title("Q3 (g) Addn Hyperplane : X1 - X2 - 0.3 = 0", color='black', fontsize=18)
plt.legend(facecolor='black')
plt.show()
```

#### Q4. This problem involves the OJ data set which is part of the ISLR package.

```
# imports
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score
```

```
# get data
oj_data = pd.read_csv("https://raw.githubusercontent.com/makbigc/ISLR/master/datasets/OJ.csv")
oj_data.drop(['Unnamed: 0'], axis=1, inplace=True)

# encoding cchar features
oj_data['Purchase'] = oj_data['Purchase'].map({'CH': 0, 'MM': 1})
oj_data['Store7'] = oj_data['Store7'].map({'No': 0, 'Yes': 1})

# features 
X = oj_data.iloc[:, 1:]
y = oj_data.iloc[:, 0]
```

```
# a)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    train_size=800,
                                                    random_state=42)
```

```
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
# b)
clf = svm.SVC(kernel='linear', C=0.01)
clf.fit(X_train, y_train.values.ravel())
```

```
print("Intecept: ", clf.intercept_)
```

```
print("Support vectors: ", clf.support_vectors_)
```

```
print("Number of support vectors : ", len(clf.support_vectors_))
```

```
print(classification_report(y_train, clf.predict(X_train)))
```

From the classification report above it can be observed that the classifier identifies 74% and 80% of the predictions correctly for both classes respectively.

Classifier is able to identify 95% of the class 0 while only 46% of the class 1.

```
# c)
clf_train_pred = clf.predict(X_train)
clf_test_pred = clf.predict(X_test)

clf_train_acc = accuracy_score(y_train, clf_train_pred)
clf_test_acc = accuracy_score(y_test, clf_test_pred)

print("Training error rate : ", 1 - clf_train_acc)
print("Test error rate : ", 1 - clf_test_acc)
```

```
# d)
params = {
    'C': [0.01, 0.05, 0.50, 1, 10]
}

model = svm.SVC(kernel = 'linear', random_state=42)

clf = GridSearchCV(model, 
                   param_grid=params,
                   n_jobs=-1,
                   verbose=2)

clf.fit(X_train, y_train.values.ravel())
```

```
print("Optimal Cost hyper-parameter : ", clf.best_estimator_)
```

```
# e)
clf_best = clf.best_estimator_

clf_train_pred = clf_best.predict(X_train)
clf_test_pred = clf_best.predict(X_test)

clf_train_acc = accuracy_score(y_train, clf_train_pred)
clf_test_acc = accuracy_score(y_test, clf_test_pred)

print("Training error rate : ", 1 - clf_train_acc)
print("Test error rate : ", 1 - clf_test_acc)
```

```
# f) b)
clf = svm.SVC(kernel='rbf', C=0.01)
clf.fit(X_train, y_train.values.ravel())

print("Intecept: ", clf.intercept_)

print("\nSupport vectors: ", clf.support_vectors_)

print("\nNumber of support vectors : ", len(clf.support_vectors_))

print("\n", classification_report(y_train, clf.predict(X_train)))
```

```
# f c)
clf_train_pred = clf.predict(X_train)
clf_test_pred = clf.predict(X_test)

clf_train_acc = accuracy_score(y_train, clf_train_pred)
clf_test_acc = accuracy_score(y_test, clf_test_pred)

print("Training error rate : ", 1 - clf_train_acc)
print("Test error rate : ", 1 - clf_test_acc)
```

```
# f d)
params = {
    'C': [0.001, 0.01, 0.05, 0.50, 1, 10]
}

model = svm.SVC(kernel = 'rbf', random_state=42)

clf = GridSearchCV(model, 
                   param_grid=params,
                   n_jobs=-1,
                   verbose=2)

clf.fit(X_train, y_train.values.ravel())
```

```
print("Optimal Cost hyper-parameter : ", clf.best_estimator_)
```

```
# f) e)

clf_best = clf.best_estimator_

clf_train_pred = clf_best.predict(X_train)
clf_test_pred = clf_best.predict(X_test)

clf_train_acc = accuracy_score(y_train, clf_train_pred)
clf_test_acc = accuracy_score(y_test, clf_test_pred)

print("Training error rate : ", 1 - clf_train_acc)
print("Test error rate : ", 1 - clf_test_acc)
```

#### Q5. MNIST

```
# imports
from sklearn.datasets import fetch_openml

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

from sklearn.decomposition import PCA
```

```
# read_data
def sort_by_target(mnist):
    reorder_train = np.array(sorted([(target, i) for i, target in enumerate(mnist.target[:60000])]))[:, 1]
    reorder_test = np.array(sorted([(target, i) for i, target in enumerate(mnist.target[60000:])]))[:, 1]
    mnist.data[:60000] = mnist.data[reorder_train]
    mnist.target[:60000] = mnist.target[reorder_train]
    mnist.data[60000:] = mnist.data[reorder_test + 60000]
    mnist.target[60000:] = mnist.target[reorder_test + 60000]

mnist = fetch_openml('mnist_784', version=1, cache=True)
mnist.target = mnist.target.astype(np.int8)
sort_by_target(mnist)
```

```
mnist["data"], mnist["target"]
```

```
X, y = mnist['data'], mnist['target']
X.shape
```

```
# create training and test set
X_train = np.concatenate([X[(y==3)][:400], X[(y==5)][:400]])
y_train = np.concatenate([y[y==3][:400], y[y==5][:400]])

X_test = np.concatenate([X[(y==3)][400:800], X[(y==5)][400:800]])
y_test = np.concatenate([y[y==3][400:800], y[y==5][400:800]])
```

```
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```
# scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.astype(np.float64))
X_test_scaled = scaler.transform(X_test.astype(np.float64))
```

```
# a)
log_clf = LogisticRegression(random_state=42)

log_clf.fit(X_train_scaled, y_train)
```

```
y_pred_train = log_clf.predict(X_train_scaled)
y_pred_test = log_clf.predict(X_test_scaled)
```

```
print("Training Clasification : \n\n")

print(classification_report(y_train, y_pred_train))
```

```
print("Testing Classification: \n\n")

print(classification_report(y_test, y_pred_test))
```

Prediction on training has been 100% accurate. While on the test set it has been fairly accurate with F1 score of 0.94 for each class.

```
print("Training Mis-classification rate: ", 1 - accuracy_score(y_train, y_pred_train))
print("Testing Mis-classification rate: ", 1 - accuracy_score(y_test, y_pred_test))
```
